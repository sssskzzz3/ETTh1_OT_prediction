```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, LSTM, Dropout, Dense
from tensorflow.keras.models import Model
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.utils import plot_model
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

df = pd.read_csv(r"C:\Users\z\Desktop\数据挖掘\RNN时间序列预测\ETTh1.csv")
# 转换日期列为 datetime 格式
df['date'] = pd.to_datetime(df['date'])
# 设置日期为索引
df.set_index('date', inplace=True)
# 选择需要的特征列
features = ['HUFL', 'HULL', 'MUFL', 'MULL', 'LUFL', 'LULL']
target = 'OT'  # 提取特征数据和目标数据
X = df[features].values
y = df[target].values
# 归一化特征和目标数据
scaler_X = MinMaxScaler(feature_range=(0, 1))
scaler_y = MinMaxScaler(feature_range=(0, 1))
X_scaled = scaler_X.fit_transform(X)
```

```python
y_scaled = scaler_y.fit_transform(y.reshape(-1, 1))
# 创建时间步数据集（使用过去 24 小时数据来预测未来 24 小时）
def create_dataset(X, y, time_step=24):
    X_data, y_data = [], []
    for i in range(len(X) - time_step):
        X_data.append(X[i:i+time_step])
        y_data.append(y[i+time_step])
    return np.array(X_data), np.array(y_data)
# 创建训练数据集
X_data, y_data = create_dataset(X_scaled, y_scaled, time_step=24)
# 将数据分为训练集和测试集
X_train, X_test, y_train, y_test = train_test_split(X_data, y_data, test_size=0.2, shuffle=False)
# 打印数据的形状
print(f'X_train shape: {X_train.shape}')
print(f'X_test shape: {X_test.shape}')
```

```python
# 定义LSTM模型
def create_optimized_lstm_model(input_shape):
    # 使用 Input 层来定义输入形状
    inputs = Input(shape=input_shape)
    
    # LSTM 层
    x = LSTM(units=128, activation='tanh', return_sequences=True)(inputs)
    x = Dropout(0.2)(x)  # 添加 Dropout 层防止过拟合
    x = LSTM(units=64, activation='tanh')(x)  # 第二个 LSTM 层
    x = Dropout(0.2)(x)  # 添加 Dropout 层
    output = Dense(1)(x)  # 输出预测值
    
    # 创建模型
    model = Model(inputs=inputs, outputs=output)
    
    model.compile(optimizer='adam', loss='mse')
    return model

# 创建 LSTM 模型
optimized_lstm_model = create_optimized_lstm_model(input_shape=(X_train.shape[1], X_train.shape[2]))
# 显示模型结构
optimized_lstm_model.summary()
# 可视化并保存模型结构图
plot_model(optimized_lstm_model, to_file='lstm_model_structure.png', show_shapes=True, show_layer_names=True)
# 早停机制
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
# 训练 LSTM 模型
history = optimized_lstm_model.fit(X_train, y_train, epochs=100, batch_size=64, validation_data=(X_test, y_test), callbacks=[early_stopping])
# 使用模型进行预测
y_pred_optimized_lstm = optimized_lstm_model.predict(X_test)
# 反归一化预测值
y_pred_optimized_lstm_rescaled = scaler_y.inverse_transform(y_pred_optimized_lstm)
# 反归一化真实值
y_test_rescaled = scaler_y.inverse_transform(y_test.reshape(-1, 1))
# 打印预测的形状
print(f'Predicted shape: {y_pred_optimized_lstm_rescaled.shape}')
print(f'Real shape: {y_test_rescaled.shape}')
# 绘制并保存 LSTM 预测结果
plt.figure(figsize=(12, 6))
# 绘制真实值
plt.plot(y_test_rescaled, label='True OT')
# 绘制优化后的 LSTM 预测值
plt.plot(y_pred_optimized_lstm_rescaled, label='LSTM Predicted OT')
plt.legend()
plt.title('LSTM Model - Prediction vs True Values')
plt.xlabel('Time')
plt.ylabel('OT')
# 保存图像
plt.savefig('LSTM预测结果', dpi=600)
plt.show()

# 计算均方误差 MSE 和均方根误差 RMSE
mse = mean_squared_error(y_test_rescaled, y_pred_optimized_lstm_rescaled)
rmse = np.sqrt(mse)
# 计算平均绝对误差 MAE
mae = mean_absolute_error(y_test_rescaled, y_pred_optimized_lstm_rescaled)

# 打印评估指标
print(f'Mean Squared Error (MSE): {mse}')
print(f'Root Mean Squared Error (RMSE): {rmse}')
print(f'Mean Absolute Error (MAE): {mae}')

# 获取训练和验证集的损失值
train_loss = history.history['loss']
val_loss = history.history['val_loss']

# 绘制损失函数曲线
plt.figure(figsize=(12, 6))
plt.plot(train_loss, label='Train Loss')
plt.plot(val_loss, label='Validation Loss')

plt.title('LSTM Model - Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

# 保存损失函数曲线
plt.savefig('LSTM_loss_curve', dpi=600)
plt.show()
```
